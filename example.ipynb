```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from tifffile import imread
import numpy as np

import colormaps as c

%matplotlib widget
```

```python
# Test Images
im = imread( "test.tif" )
xx,yy = np.meshgrid( np.linspace(0,1,256),np.linspace(0,1,256))
```

```python
# List of all cmaps
cmaps = [c.iris, c.purple_wing, c.chroma_cardioid, c.copper_5N, c.cryo, c.crayon_box,c.lapis,c.planckian_locus]
cmap_label = ['iris','purple_wing','chroma_cardiod','copper_5N','cryo','crayon_box','lapis','planickan_locus']
N = len(cmaps)
```

```python
x = np.linspace(0,1,256)
fig, ax = plt.subplots(N,2, figsize=(3,6))

for i in range(N):
    cmap = ListedColormap(cmaps[i](x), name="my_cmap")
    ax[i,0].pcolormesh( xx, cmap=cmap )
    ax[i,0].axis('scaled')
    ax[i,0].axis('off')
    ax[i,0].set_title(cmap_label[i])
    ax[i,1].pcolormesh( im, cmap=cmap )
    ax[i,1].axis('scaled')
    ax[i,1].axis('off')
```
